```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Carregar os DataFrames no EDA - Individual dos Dados

```python
# Carregamento do eda do Sensor 1 - eda_dados_1
df_carregamento_1 = pd.read_csv('./datasets/eda_dados_1.csv')
```

```python
# Informações do Sensor 1
df_carregamento_1.info()
```

```python
# Carregamento do eda do Sensor 2 - eda_dados_2
df_carregamento_2 = pd.read_csv('./datasets/eda_dados_2.csv', skiprows=0)
```

```python
# Informações do Sensor 2
df_carregamento_2.info()
```

```python
# Carregamento do eda do Sensor 3 - eda_dados_3
df_carregamento_3 = pd.read_csv('./datasets/eda_dados_3.csv', skiprows=0)
```

```python
# Informações do Sensor 3
df_carregamento_3.info()
```

```python
# Informações do Sensor 3 
df_carregamento_3.info()
```

```python
# Carregamento do eda do Sensor 4 - eda_dados_4
df_carregamento_4 = pd.read_csv('./datasets/eda_dados_4.csv', skiprows=0)
```

```python
# Informações do Sensor 4
df_carregamento_4.info()
```

```python
# Carregamento do eda do Sensor 5 - eda_dados_5
df_carregamento_5 = pd.read_csv('./datasets/eda_dados_5.csv', skiprows=0)
```

```python
# Informações do Sensor 5
df_carregamento_5.info()
```

```python
# Unificando os datasets
df_sensores = pd.concat([df_carregamento_1, df_carregamento_2, df_carregamento_3, df_carregamento_4, df_carregamento_5], axis=0, ignore_index=True)
```

```python
# Informações do Dataset Resultante da Concatenação
df_sensores.info()
```

# EDA

```python
# Medidas Estatísticas das variáveis
df_sensores.describe()
```

```python
# Verificar se temos outliers
# Plot BoxPlot
sns.boxplot(df_sensores, y='0')
```

```python
sns.scatterplot(data=df_sensores, x='Classe', y='0')
```

```python
# Verificar Correlação - Pearson
sns.heatmap(df_sensores.corr('pearson'), annot = True)
```

```python
# Verificar Correlação - Pearson
sns.heatmap(df_sensores.corr('pearson'), vmin=1, vmax=1, annot = True)
```

```python
# Mapa de Correlação só com a variável Target (Classe)
sns.heatmap(df_sensores.corr()[['Classe']].sort_vales(by='Classe', ascending=False), vmin=1, vmax=1, annot=True, cmap='BrBG')
```

```python
# Verificar Correlação - Spearman (Relação ao lineares)
sns.heatmap(df_sensores.corr('spearman'), annot = True)
```

```python
# Histograma das Variáveis
sns.displot(df_sensores, x='0')
```

```python
# Histograma de todas as variáveis
df_sensores.hist(layout={2,3}, figsize={10,5})
```

```python
# Histogramas das variáveis
sns.pairplot(df_sensores)
```
