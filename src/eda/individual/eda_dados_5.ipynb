```python
import numpy as np
import pandas as pd
```

# Criar DataFrames com base no Dataset Dados_5

```python
# Carregar os dados do arquivo .npy
dados = np.load('../../datasets/Classes.npy', allow_pickle=True)

# Converter os dados em um DataFrame pandas
classes = pd.DataFrame(dados)

# Alterar o nome da primeira coluna para "Classe"
classes = classes.rename(columns={classes.columns[0]: "Classe"})

# Exibir o DataFrame
print(classes)
```

```python
# Carregar os dados .npy usando o NumPy
dados5_numpy = np.load('../../datasets/Dados_5.npy')

# Converter os dados para um DataFrame pandas
df_carregamento_5 = pd.DataFrame(dados5_numpy)
```

```python
# Mostrar 5 primeiros registros do Dataframe
df_carregamento_5.head(5)
```

```python
# Mostrar 5 Ãºltimos registros
df_carregamento_5.tail(5)
```

```python
# Mostrar estrutura / schema do Dataframe
df_carregamento_5.info()
```

```python
# Unificando a Classe com o Dataframe 
df_dados_5 = pd.concat([classes, df_carregamento_5], axis=1)
print(df_dados_5)
```

```python
# Mostrar estrutura / schema do Dataframe
df_dados_5.info()
```

# TransformaÃ§Ã£o de Dados

```python
pd.set_option('future.no_silent_downcasting', True)

# Mapear os valores para os nÃºmeros desejados
mapeamento = {'Classe A': 1, 'Classe B': 2, 'Classe C': 3, 'Classe D': 4, 'Classe E': 5}

# Substituir os valores na primeira coluna de acordo com o mapeamento
df_dados_5.iloc[:, 0] = df_dados_5.iloc[:, 0].replace(mapeamento)
```

```python
# Mostrar 5 primeiros registros do Dataframe
df_dados_5.head(5)
```

```python
# Mostrar 5 Ãºltimos registros
df_dados_5.tail(5)
```

# DetecÃ§Ã£o de Valores Ausentes

```python
# Quantidade de colunas que possuem pelo menos 1 valor ausente NaN
df_dados_5.isna().any(axis=0).sum()
```

```python
# Detectar valores ausentes em todas as colunas do Dataset (Qtd)
df_dados_5.isna().sum()
```

# Remover Valores Ausentes

```python
# Remover linhas com TODOS os valores ausentes das colunas
df_dados_5.dropna(axis=0, how='all', inplace=True)
```

```python
# Remover linhas com algum valor ausente nas colunas
df_dados_5.dropna(axis=0, inplace=True)
```

```python
# Quantidade de colunas que possuem pelo menos 1 valor ausente NaN
df_dados_5.isna().any(axis=0).sum()
```
