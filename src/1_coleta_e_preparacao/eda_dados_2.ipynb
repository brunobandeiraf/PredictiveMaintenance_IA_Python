```python
import numpy as np
import pandas as pd
```

## Criar DataFrames com base no Dataset Dados_2

```python
# Carregar os dados do arquivo .npy
dados = np.load('./datasets/Classes.npy', allow_pickle=True)

# Converter os dados em um DataFrame pandas
classes = pd.DataFrame(dados)

# Exibir o DataFrame
print(classes)
```

```python
# Carregar os dados .npy usando o NumPy
dados2_numpy = np.load('./datasets/Dados_2.npy')

# Converter os dados para um DataFrame pandas
df_carregamento_2 = pd.DataFrame(dados2_numpy)
```

```python
# Mostrar 5 primeiros registros do Dataframe
df_carregamento_2.head(5)
```

```python
# Mostrar 5 Ãºltimos registros
df_carregamento_2.tail(5)
```

```python
# Mostrar estrutura / schema do Dataframe
df_carregamento_2.info()
```

```python
# Unificando a Classe com o Dataframe 
df_dados_2 = pd.concat([classes, df_carregamento_2], axis=1)
print(df_dados_2)
```

```python
# Mostrar estrutura / schema do Dataframe
df_dados_2.info()
```

# TransformaÃ§Ã£o de Dados

```python
pd.set_option('future.no_silent_downcasting', True)

# Mapear os valores para os nÃºmeros desejados
mapeamento = {'Classe A': 1, 'Classe B': 2, 'Classe C': 3, 'Classe D': 4, 'Classe E': 5}

# Substituir os valores na primeira coluna de acordo com o mapeamento
df_dados_2.iloc[:, 0] = df_dados_2.iloc[:, 0].replace(mapeamento)
```

```python
# Mostrar 5 primeiros registros do Dataframe
df_dados_2.head(5)
```

```python
# Mostrar 5 Ãºltimos registros
df_dados_2.tail(5)
```

```python
# Removendo a Ãºltima coluna do Dataframe
df_dados_2.drop(df_dados_2.columns[-1], axis=1, inplace=True)
```
